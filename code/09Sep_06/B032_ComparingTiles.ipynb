```python
from shapely.geometry import Point
```

```python
p_centroid = Point([475000,175000])
```

```python
p = db.contains('pixels', p_centroid)
```

```python
p_tiles1 = db.intersects('tiles1', p.geometry[0].buffer(1000))
p_tiles2 = db.intersects('tiles2', p.geometry[0].buffer(1000))
p_tiles3 = db.intersects('tiles3', p.geometry[0].buffer(1000))
p_tiles4 = db.intersects('tiles4', p.geometry[0].buffer(1000))
```
