```python
api = 'https://api.beta.ons.gov.uk/v1/datasets'
```

```python
r = requests.get(api)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(r.json()['items'])
```

```python
df.id.unique()
```

```python
df.href= ''
l = list()
for i in range(len(df)):
    href = df.iloc[i].links['latest_version']['href']
    l.append(href)
```

```python
df['href'] = pd.Series(l)
```

```python
df['get'] = df['href'].apply(lambda x: requests.get(x))
```

```python
df['durl'] = df['get'].apply(lambda x: x.json()['downloads']['csv']['href'])
```

```python
import os
```

```python
def download_data(row):
    fpath = f'../data/ons/{row.id}'
    if not os.path.exists(fpath):
        os.mkdir(fpath)
    
    with open(f'../data/ons/{row.id}/data.csv','wb') as f:
        f.write(requests.get(row.durl).content)
        f.close()
```

```python
df.apply(download_data, axis=1)
```
